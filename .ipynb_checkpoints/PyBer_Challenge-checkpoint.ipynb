# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
city_data_df.head()
```

```python
ride_data_df.head()
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_df = pyber_data_df.groupby("type").count()["ride_id"]
total_rides_df
```

```python
# 2. Get the total drivers for each city type
total_drivers_df = city_data_df.groupby("type").sum()["driver_count"]
total_drivers_df
```

```python
#  3. Get the total amount of fares for each city type
total_fares_df = pyber_data_df.groupby("type").sum()["fare"]
total_fares_df
```

```python
#  4. Get the average fare per ride for each city type.
average_fare_per_ride_df = total_fares_df / total_rides_df
average_fare_per_ride_df
```

```python
# 5. Get the average fare per driver for each city type.
average_fare_per_driver_df = total_fares_df / total_drivers_df
average_fare_per_driver_df
```

```python
#  6. Create a PyBer summary DataFrame.
summary_df = pd.DataFrame()

summary_df["Total Rides"] = total_rides_df
summary_df["Total Drivers"] = total_drivers_df
summary_df["Total Fares"] = total_fares_df
summary_df["Average Fare per Ride"] = average_fare_per_ride_df
summary_df["Average Fare per Driver"] = average_fare_per_driver_df

summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
summary_df.index.name = None
summary_df
```

```python
#  8. Format the columns.
summary_df["Total Rides"] = summary_df["Total Rides"].map("{:,}".format)
summary_df["Total Drivers"] = summary_df["Total Drivers"].map("{:,}".format)
summary_df["Total Fares"] = summary_df["Total Fares"].map("${:,.2f}".format)
summary_df["Average Fare per Ride"] = summary_df["Average Fare per Ride"].map("${:.2f}".format)
summary_df["Average Fare per Driver"] = summary_df["Average Fare per Driver"].map("${:.2f}".format)

summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
city_fares = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
city_type_df = pd.DataFrame(city_fares)
city_type_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
city_type_df = city_type_df.reset_index()
city_type_df.head(10)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date.
city_type_pivot = city_type_df.pivot(index="date", columns="type", values="fare")
city_type_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
Jan_April_fares_df = city_type_pivot.loc['2019-01-01':'2019-04-28']
Jan_April_fares_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
Jan_April_fares_df.index = pd.to_datetime(Jan_April_fares_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
Jan_April_fares_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
weekly_fares_df = Jan_April_fares_df.resample("W").sum()
weekly_fares_df.head(10)
```

```python
weekly_fares_df.tail(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
weekly_fares_df.plot(figsize=(25,10))
plt.title("Total Fare by City Type", fontsize=30)
plt.ylabel("Fare ($USD)", fontsize=20)
plt.xlabel("Month", fontsize=20)
lgnd = plt.legend(fontsize="17", loc="center", title="type")
lgnd.get_title().set_fontsize(17)
plt.tight_layout()
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
